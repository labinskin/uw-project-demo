```python
import pandas as pd

df = pd.read_csv('Offenses_by_Agency_Table_data_WI_2018data.csv')
df2=pd.read_csv('Offenses_by_Agency_Table_data_WI_2019data.csv')
df4=pd.read_csv('Offenses_by_Agency_Table_data_WI_2020data.csv')

df2.head() 
```

```python
df3=pd.merge(df,df2,on=['Agency Name','Offense Description'])
df3.head(25)
```

```python
df5=pd.merge(df3,df4,on=['Agency Name','Offense Description'])
df5.head(25)
```

```python
df5.to_csv('2018_2019_2020.csv')
```
