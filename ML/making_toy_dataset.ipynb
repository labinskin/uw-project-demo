```python
# Import Dependencies
import pandas as pd
```

```python
# Make a reference to the books.csv file path
csv_path = "../Data/alcohol_consumption.csv"

# Import the books.csv file as a DataFrame
alcohol_con_df = pd.read_csv(csv_path)
alcohol_con_df
```

```python
# Add prefix to denote alcohol consumption drinking
alcohol_con_df=alcohol_con_df.add_prefix('alcohol_consumption_')
#df = df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'})
alcohol_con_df=alcohol_con_df.rename(columns={'alcohol_consumption_State': 'State', 'alcohol_consumption_County': 'County'})
alcohol_con_df.head()
```

```python
# Make a reference to the books.csv file path
csv_path = "../Data/binge_drinking.csv"

# Import the books.csv file as a DataFrame
binge_drinking_df = pd.read_csv(csv_path)
binge_drinking_df
```

```python
# Merge two dataframes using an inner join (default).
merge_df = pd.merge(alcohol_con_df, binge_drinking_df, on="County")
merge_df=merge_df.rename(columns={'State_x': 'State'})
merge_df
```

```python
# Make a reference to the books.csv file path
csv_path = "../Data/offenses_2018.csv"

# Import the books.csv file as a DataFrame
offenses_2018_df = pd.read_csv(csv_path)
offenses_2018_df
```

```python
# Make a reference to the books.csv file path
csv_path = "../Data/offenses_2019.csv"

# Import the books.csv file as a DataFrame
offenses_2019_df = pd.read_csv(csv_path)
offenses_2019_df
```

```python
# Concat the two offenses data frames
offenses_merge_df = pd.concat([offenses_2018_df, offenses_2019_df], ignore_index=True)
offenses_merge_df
```

```python
# Make a reference to the books.csv file path
csv_path = "../Data/offenses_2020.csv"

# Import the books.csv file as a DataFrame
offenses_2020_df = pd.read_csv(csv_path)
offenses_2020_df
```

```python
# Concat the 2020 offenses data frame to the merged
offenses_merge_df = pd.concat([offenses_merge_df, offenses_2020_df], ignore_index=True)
offenses_merge_df
```

```python
#Now for the big one!!!
total_merge_df = pd.merge(offenses_merge_df, merge_df, on='County', how='left')
total_merge_df
```
